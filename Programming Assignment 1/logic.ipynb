```python
def resolution(cnf: Set[Set[Predicate]]) -> bool:
    # INPUT CRITERIA:
    # 1. No existential quantifiers
    # 2. No free variables
    # 3. All quantifiers must come at beginning
    # 4. Is in CNF
        # a. Negation can be applied to atomic formulas
        # b. Formulas have a max depth of 2
#     print("cnf", cnf)
    while True:
        for s in list(cnf):
            for s2 in list(cnf):
                if s != s2:
                    for c in list(s):
                        for c2 in list(s2):
                            if "¬{}".format(str(c2).split("(")[0]) == str(c).split("(")[0] or "¬{}".format(str(c).split("(")[0]) == str(c2).split("(")[0]:
                                newset = set(s)
                                newset2 = set(s2)
                                newset.remove(c)
                                newset2.remove(c2)
                                if len(newset) == 0:
                                    final_new_set = frozenset(newset2)
                                elif len(newset2) == 0:
                                    final_new_set = frozenset(newset)
                                else:
                                    final_new_set = frozenset(newset.union(newset2))
                                cnf.add(final_new_set)
#                                 print('final_new_set', final_new_set)
                                if (final_new_set) == None or len(final_new_set) == 0:
                                    return False

        
    return True
```
